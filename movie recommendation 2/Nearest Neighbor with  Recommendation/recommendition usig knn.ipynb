```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
movie_df=pd.read_csv(r"C:\Users\Acer.DESKTOP-TSARLU6\Downloads\movies (1).csv")
rating_df = pd.read_csv(r"C:\Users\Acer.DESKTOP-TSARLU6\Downloads\ratings.csv")
rating_df
```

```python
movie = movie_df.merge(rating_df,on='movieId')
```

```python
movie.drop(columns=['genres','timestamp'],inplace=True)
```

```python
movie
```

```python
combine_movie= movie.dropna(axis=0,subset=['title'])
combine_movie
```

```python
total_rating_count=combine_movie.groupby('title')['rating'].count().reset_index().rename(columns={'rating':'total rating'})
```

```python
rating_with_count=combine_movie.merge(total_rating_count,on='title')
```

```python
pd.set_option('display.float_format', lambda x: '%.3f' % x)
total_rating_count.describe()
```

```python
popularity_thershold=50
```

```python
rating_with_count
```

```python
popular_movies=rating_with_count[rating_with_count['total rating']>=popularity_thershold]
```

```python
rating_with_count.query('`total rating` >= @popularity_thershold')
```

```python
popular_movies.shape
```

```python
moviemat = popular_movies.pivot_table(index='title',columns='userId',values='rating').fillna(0)
moviemat
```

```python
from scipy.sparse import csr_matrix
moviemat_matrix=csr_matrix(moviemat)
moviemat_matrix
```

```python
from sklearn.neighbors import NearestNeighbors
knn = NearestNeighbors(metric='cosine',algorithm='brute')
knn.fit(moviemat_matrix)
```

```python
query_idex=np.random.choice(moviemat.shape[0])
print(query_idex)
distance,indices=knn.kneighbors(moviemat.iloc[query_idex,:].values.reshape(1,-1),n_neighbors=6)
print(distance)
print(indices)
```

```python
def recommend_movie(query_index):
    distances,indices=knn.kneighbors(moviemat.iloc[query_index,:].values.reshape(1,-1),n_neighbors=6)
    for i in range(0,len(distances.flatten())):
        if i==0:
            print('Recommendation for {0}'.format(moviemat.index[query_index]))
        else:
            print('{0}:{1} with the distance of {2}'.format(i,moviemat.index[indices.flatten()[i]],distances.flatten()[i]))
```

```python
recommend_movie(174)
```
